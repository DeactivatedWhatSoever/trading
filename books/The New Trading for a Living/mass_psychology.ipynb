## Strong Feelings

More buyers than sellers? Supply and demand says this. If there are people who 'want' to buy more, there's going to be a price raise. People who want to make more, will make the price go down.

The buys and sells never change in number. There's no such thing as more buyers than sellers. If 200 shares are sold, 200 shares are bought. You have to think that it's just a change in that price. 

To make the trend go up, the bulls must have optimistic feelings about the price. By optimistic, they'll feel that the price will go up and up. The stronger the feelings, the sharper the price will go up in a given timeframe. Also, the bears must be scared to sell the shares in a high price. They'll keep trying and trying not to sell at low price. They'll keep on waiting for the highest price until the bulls lose enthusiasm. If the bulls start to get bored, the bears come to get to their feelings. The more the bears emotions go stronger, the trend will go down. They will sell short and keep making the prices go down and profit. The bears will keep on thinking that it's a discount and keep buying until their emotions become about equal. They'll fight with the price and the market will go sideways.
