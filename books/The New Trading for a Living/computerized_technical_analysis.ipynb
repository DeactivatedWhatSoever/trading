# Moving Average Convergence Divergence: MACD Lines and MACD-Histogram

An indicator that gives signals when convergence or divergence occurs. The crossover between three EMAs. 

## How to Create MACD

The MACD has three lines. Two lines of EMAs which are solid lines and another EMA line which is dashed. Two short-term signalers, and a base of a long-term EMA. Buy and sell signals occur when the MACD lines go over or under the signal line. When it goes above, it means it might go into an uptrend. Which is a great place to buy? But when it goes under the signal line, it's best to sell since it might be on a downtrend.

How to calculate MACD?

1. Calculate a 12-day EMA of closing prices
2. Calculate a 26-day EMA of closing prices
3. Subtract the 26-day EMA from the 12-day EMA, and plot their difference as a solid line. This is the fast MACD line.
4. Calculate a 9-day EMA of the fast line, and plot the result as a dashed line. This is the slow Signal line.

## Market Psychology

When the MACD line rises above the signal line, it means the bulls are dominant. And when the MACD line goes below the signal line, it means the bears are dominant. So buy when it rises the signal line and sell when it goes below the signal line.

## Trading Rules for MACD Lines

1. When the fast MACD line crosses above the slow Signal line, it gives a buy signal. Go long, and place a protective stop below the latest minor low.
2. When the fast line crosses below the slow line, it gives a sell signal, Go short, and place a protective stop above the latest minor high.

## More on MACD Lines

# Directional System

It helps traders to see whether it's worth following the trend. The trend must move fast 'enough' to make it worth. ## How to Construct the Directional System

Directional movement is the largest part of today's range that is outside of yesterday's range. 

1. If today's range extends above yesterday's range, Directional Movement is positive (+DM)
2. If today's range extends below yesterday's range, Directional Movement is negative (-DM)
3. If today's range is inside of yesterday's range or extends above and below it by equal amounts, there is no Directional Movement (DM = 0). If today's range extends both above and below yesterday's range, DM is positive or negative, pending on which part of the "outside range" is larger.
4. On a limit-up day, +DM equals the distance from today's close to yesterday's high. On a limit-down day, -DM equals the distance from today's close to yesterday's low.

## Crowd Behavior

Directional system indicates how much power the bulls and bears have. That power comes from price movements. ## Trading Rules

2. When ADX declines, it shows that the market is becoming less directional. There are likely to be many whipsaws, When ADX points down, it is better not to use a trend-following method.
3. When ADX falls below both Directional lines, it identifies a flat, sleepy market. Do not use a trend-following system but get ready to trade, because major trends emerge from such lulls.

It shows that a new bull market or bear market is being born, depending on what Directional line is on top.
5. When ADX rallies above both Directional lines, it identifies an overheated market. When ADX turns down from above both Directional lines, it shows that the major trend has stumbled. It is a good time to take profits on a directional trade. If you trade large positions, you definitely want to take partial profits.

# Stochastic

Stochastic tracks the relationship of each closing price and the recent high-low range. It has a fast line called %K and a slow line called %D. Of course the signal will come when there'll be a divergence. This does calculate the price, and it's about comparing two prices with each other so I don't think it gives you the idea of a trend. And when there's two lines of prices, and it doesn't care about the current price, then it's about whether it's going to reverse. I've got to understand all the components between oscillators and trend-following indicators. 

1. Raw Stochastic = %
%K = (Ctod - Ln / Hn - Ln) * 100

Ctod = today's close
Ln = the lowest point for the selected number of days.
Hn = the highest point for the selected number of days
n = the number of days for Stochastic, selected by the trader.
2. %D, smoothing out %K. To make it the slow line.
%D = (3-day sum of (Ctod - Ln) / 3-day sum of (Hn - Ln)) * 100

There are two ways to plot stochastic. The fast and slow version. You see, the fast one is just plotting the one above. It leads to lots of whipsaws which is pretty much for a day trader thing. Professional traders tend to use the slow stochastic. Which is, taking the %D of the fast stochastic and having that as %K. And just calculate the %D with the %D of the slow stochastic. It gives you bigger and meaningful turns. It also filters out market noise and fewer whipsaws.

Stochastic is a percentage scale, which is 0~100. 20% and 80% are the reference marks to indicate whether it's overbought or oversold.
